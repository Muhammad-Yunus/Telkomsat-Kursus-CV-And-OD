### 4.2.1 What is YOLOv11?
- <font color="orange">YOLOv11</font> is a new state-of-the-art computer vision model built by [Ultralytics](https://www.ultralytics.com/), the creators of <font color="orange">YOLOv5</font>. 
- The <font color="orange">YOLOv11</font> model contains out-of-the-box support for <font color="orange">object detection</font>, <font color="orange">classification</font>, and <font color="orange">segmentation tasks</font>, accessible through a Python package as well as a command line interface.<br>
<img src="resource/yolo11-1.png" width="400px">

- <font color="orange">Yolo Model</font> can be downloaded from <font color="orange">Ultralytics Assets</font> repository : https://github.com/ultralytics/assets/releases/
- All model released in [Pytorch](https://pytorch.org/) Model format (`.pt`)
- For example if we want to use Yolo v11, there is several naming convention like the following,
- Based on Model Size :
    - `yolov11n.pt` : *Nano*,
    - `yolov11s.pt` : *Small*,
    - `yolov11m.pt` : *Medium*,
    - `yolov11b.pt` : *Balanced*,
    - `yolov11l.pt` : *Large*,
    - `yolov11x.pt` : *Extra Large*,
- Additional model with special capability :
    - `yolov11l-cls.pt` : *Classification Model*
    - `yolov11l-human.pt` : *Human Detection Model*
    - `yolov11l-obb.pt` : *Oriented Bounding Box Object Detection*
    - `yolov11l-pose.pt` : *Pose Estimation Model*
    - `yolov11l-seg.pt` : *Segmentation Model*<br>
