## 4.3 Object Detection Inferencing

### 4.3.1 <font color="orange">Training vs Inferencing</font> Deep Learning Model
- **Training** memerlukan **dataset** dan akan menghasilkan **Model**.
- **Inference** memerlukan **Model** dan **data test** yang akan menghasilkan **Prediction**.  
<img src="resource/training-inferencing.jpg" style="width:700px"></img><br><br>

- sample code inferencing using ultralytics
    - please download `yolo11n.pt` model from https://github.com/ultralytics/assets/releases/
    - save into /Kursus `Object Detection/Pertemuan_4` folder

```python
# Import the Ultralytics YOLO module
from ultralytics import YOLO

# import OpenCV
import cv2
```

```python
# Load a pre-trained YOLOv11 model (you can also specify a custom model path)
model = YOLO("yolo11n.pt")  
```

```python
# Run inference on an image
results = model("image1.jpg")  # path to your input image

result = results[0]  # get the first result, since we only predict for single image input

# Print results
print(result)  # prints summary of detected objects
```

```python
# You can also access detailed info
print(result.boxes)  # bounding boxes
```

```python
# Iterate through detected boxes
for box, cls, conf in zip(result.boxes.xyxy, result.boxes.cls, result.boxes.conf):

    x1, y1, x2, y2 = map(int, box)       # bounding box coordinates
    
    class_name = model.names[int(cls)]   # class label
    
    score = float(conf)                  # confidence
    
    label = f"{class_name} {score:.2f}"
    
    print(f"Box: {x1},{y1},{x2},{y2} Label: {label}")
```

```python
# Draw bounding boxes on the original image using the built-in plot method
img_with_boxes = result.plot()  
```

```python
type(img_with_boxes)
```

```python
# Show image in a window using OpenCV
cv2.imshow("YOLOv11 Inferencing", img_with_boxes)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Save the visualized image
result.save("output.jpg")
```
