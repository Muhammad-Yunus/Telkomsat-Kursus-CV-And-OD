## 6.3 Test Inferencing Custom Model

- sample code inferencing using ultralytics
    - make sure you already download `yolo11s-seg-custom.pt` from notebook 6.2 in Google Colab
    - save into /Kursus `Object Detection/Pertemuan_6` folder

```python
# Import the Ultralytics YOLO module
from ultralytics import YOLO

# import OpenCV
import cv2
```

```python
# Load a pre-trained YOLOv11 Segmentation model (you can also specify a custom model path)
model = YOLO("yolo11s-seg-custom.pt")  
```

```python
# Set confidence threshold, below which detections will be ignored
CONF_THRESHOLD = 0.5
```

```python
# Run inference on an image
results = model("pohon_test1.jpg")  # path to your input image

result = results[0]  # get the first result, since we only predict for single image input

# Filter boxes by confidence
# only keep boxes with confidence >= CONF_THRESHOLD
mask = result.boxes.conf >= CONF_THRESHOLD
result.boxes = result.boxes[mask]


# Iterate through detected boxes
for box, cls, conf in zip(result.boxes.xyxy, result.boxes.cls, result.boxes.conf):

    x1, y1, x2, y2 = map(int, box)       # bounding box coordinates
    
    class_name = model.names[int(cls)]   # class label
    
    score = float(conf)                  # confidence
    
    label = f"{class_name} {score:.2f}"
    
    print(f"Box: {x1},{y1},{x2},{y2} Label: {label}")


# Draw bounding boxes on the original image using the built-in plot method
img_with_boxes = result.plot()  


# Show image in a window using OpenCV
cv2.imshow("YOLOv11 Inferencing", img_with_boxes)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
