## 1.4.1 Colorspace Conversion
![](res/gray_image.png)
- function `cv2.cvtColor(img, flag)` 
- parameter :
    - `img` : image matrix (numpy array)
    - `flag` :
        - convert BGR <--> RGB \
        `cv2.COLOR_BGR2RGB` \
        `cv2.COLOR_RGB2BGR`<br><br>
        - convert BGR <--> GRAY \
        `cv2.COLOR_BGR2GRAY` \
        `cv2.COLOR_GRAY2RGB` <br><br>
- Convert BGR to RGB illustration. <br><br>
    <img src="res/gray_image_2.png" style="width: 600px;background-color:white"></img><br><br>

## 4.1 OpenCV Implementation

```python
# read image from file 'sample1.tif'
img = cv2.imread('sample1.tif')


# convert BGR to Gray
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow('Original', img)
cv2.imshow('Gray', img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
